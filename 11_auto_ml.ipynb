````python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
import optuna
from sklearn.datasets import fetch_california_housing
````

````python
data = fetch_california_housing()
data
````

````python
X_data = data.data
y_data = data.target

X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=42)
````

````python
def objective(trial) :
    model_type = trial.suggest_categorical("model", ["RandomForest", "GradientBoost", "XGBoost"])

    if model_type == "RandomForest" :
        model = RandomForestRegressor(                                  # GPT 한테 물어보기 : 하이퍼파라미터 뭐가 있는지, 어떤 의미인지
            n_estimators=trial.suggest_int("n_estimators", 100, 1500),  # optuna한테 100부터 1500 사이의 값 중 최적화된 값 찾아줘
            max_depth=trial.suggest_int("max_depth", 3, 50),
            min_samples_split=trial.suggest_int("min_samples_split", 2, 10),
            random_state = 42
        )
    elif model_type == "GradientBoost" :
        model = GradientBoostingRegressor(
            n_estimators=trial.suggest_int("n_estimators", 100, 1500),
            max_depth=trial.suggest_int("max_depth", 3, 50),
            learning_rate=trial.suggest_float("learning_rate", 0.01, 0.3),
            random_state = 42
        )
    elif model_type == "XGBoost" :
        model = XGBRegressor(
            n_estimators=trial.suggest_int("n_estimators", 100, 1500),
            max_depth=trial.suggest_int("max_depth", 3, 50),
            learning_rate=trial.suggest_float("learning_rate", 0.01, 0.3),
            subsample=trial.suggest_float("subsample", 0.5, 1.0),
            random_state = 42
        )
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    return r2_score(y_test, y_pred)
````

````python
# direction : maximize -> 결과값이 커지는 방향으로 하이퍼파라미터 튜닝 (: r2 score)
# direction : minimiz -> 작아지는 방향 (: mse, mae)
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=50)

print(f"최고의 모델 : {study.best_params['model']}")
print(f"최고의 파라미터 : {study.best_params}")
print(f"최고 스코어 : {study.best_value}")
````

````python
# 최고의 모델 : XGBoost -> 결과 나온거 하이퍼파라미터 넣어서 다시 돌려보기
model = XGBRegressor(
            n_estimators=669,
            max_depth=5,
            learning_rate=0.07618496173623163,
            subsample=0.8890463509673252,
            random_state = 42
        )
model.fit(X_train, y_train)
````

````python
# 모델 저장하기
import joblib
joblib.dump(model, "model/best_XGBoost_Regressor.pkl")
````

````python
# 모델 불러오기
loaded_model = joblib.load("model/best_XGBoost_Regressor.pkl")

# 예측 사용
predictions = loaded_model.predict(X_test)
````

## 모델 저장하기

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import joblib

# 데이터 준비 및 모델 학습
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y)
model = RandomForestClassifier()
model.fit(X_train, y_train)

# 모델 저장
joblib.dump(model, 'model/model.pkl')
````
