````python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
import optuna
from sklearn.datasets import fetch_california_housing
````

````python
data = fetch_california_housing()
data
````

````python
X_data = data.data
y_data = data.target

X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=42)
````

````python
def objective(trial) :
    model_type = trial.suggest_categorical("model", ["RandomForest", "GradientBoost", "XGBoost"])

    if model_type == "RandomForest" :
        model = RandomForestRegressor(                                  # GPT   :   ,  
            n_estimators=trial.suggest_int("n_estimators", 100, 1500),  # optuna 100 1500      
            max_depth=trial.suggest_int("max_depth", 3, 50),
            min_samples_split=trial.suggest_int("min_samples_split", 2, 10),
            random_state = 42
        )
    elif model_type == "GradientBoost" :
        model = GradientBoostingRegressor(
            n_estimators=trial.suggest_int("n_estimators", 100, 1500),
            max_depth=trial.suggest_int("max_depth", 3, 50),
            learning_rate=trial.suggest_float("learning_rate", 0.01, 0.3),
            random_state = 42
        )
    elif model_type == "XGBoost" :
        model = XGBRegressor(
            n_estimators=trial.suggest_int("n_estimators", 100, 1500),
            max_depth=trial.suggest_int("max_depth", 3, 50),
            learning_rate=trial.suggest_float("learning_rate", 0.01, 0.3),
            subsample=trial.suggest_float("subsample", 0.5, 1.0),
            random_state = 42
        )
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    return r2_score(y_test, y_pred)
````

````python
# direction : maximize ->      (: r2 score)
# direction : minimiz ->   (: mse, mae)
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=50)

print(f"  : {study.best_params['model']}")
print(f"  : {study.best_params}")
print(f"  : {study.best_value}")
````

````python
#   : XGBoost ->      
model = XGBRegressor(
            n_estimators=669,
            max_depth=5,
            learning_rate=0.07618496173623163,
            subsample=0.8890463509673252,
            random_state = 42
        )
model.fit(X_train, y_train)
````

````python
#  
import joblib
joblib.dump(model, "model/best_XGBoost_Regressor.pkl")
````
