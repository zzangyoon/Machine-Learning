```python
# VIF(다중공선성) 확인
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
import pandas as pd

# 예: 수치형 변수만 선택
X = df[["enginesize", "horsepower", "carwidth", "curbweight"]]

# 상수항 추가
X = add_constant(X)

# VIF 계산
vif = pd.DataFrame()
vif["변수"] = X.columns
vif["VIF 값"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif)
```

```python
cols = ["symboling", "CarName", "fueltype", "aspiration", "doornumber", "carbody", "drivewheel", "carwidth", "curbweight", "enginetype", "cylindernumber", "enginesize", "horsepower"]

df[cols].info()
```

```python
# CarName 종류 확인
df["CarName"].unique()
```

```python
# CarName 데이터 전처리 (제조사만 남기기)
df["CarName"] = df["CarName"].str.split(" ").str[0]
```

```python
df["CarName"].unique()
```

```python
df["CarName"] = df["CarName"].replace("toyouta", "toyota")
```

```python
df["CarName"] = df["CarName"].replace("vokswagen", "volkswagen")
df["CarName"] = df["CarName"].replace("vw", "volkswagen")
```

```python
df["CarName"] = df["CarName"].replace("Nissan", "nissan")
```

```python
df["CarName"].unique()
```

```python
# 범주형 변수 인코딩
# "CarName" -> 라벨인코딩
# "fueltype", "aspiration", "doornumber", "carbody", "drivewheel", "enginetype", "cylindernumber" -> 원핫인코딩
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_CarName = LabelEncoder()
df["CarName_encoded"] = le_CarName.fit_transform(df["CarName"])
df
```

```python
# 원핫인코딩
features = ["fueltype", "aspiration", "doornumber", "carbody", "drivewheel", "enginetype", "cylindernumber"]
for feature in features :
    le = LabelEncoder()
    le_result = le.fit(df[feature])
    df[feature] = le_result.transform(df[feature])      # fit_transform 으로 한번에 가능
df.head()
```

```python
df = pd.get_dummies(df, columns=["fueltype", "aspiration", "doornumber", "carbody", "drivewheel", "enginetype", "cylindernumber"])
df
```

```python
df.columns
```

```python
drop_cols = ["car_ID", "CarName", "enginelocation", "carheight", "wheelbase", "carlength", "fuelsystem", "boreratio", "stroke", "compressionratio", "peakrpm", "citympg", "highwaympg", "price"]
X_data = df.drop(drop_cols, axis=1)
y_data = df["price"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
# 6. 모델 학습 함수
from sklearn.metrics import classification_report

def train_model(model) :
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(classification_report(y_pred, y_test))
```

```python
# 릿지
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

model = Ridge(alpha=0.5)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(mean_squared_error(y_pred, y_test))
print(mean_absolute_error(y_pred, y_test))
print(r2_score(y_test, y_pred))
```

```python
# 선형
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(mean_squared_error(y_pred, y_test))
print(mean_absolute_error(y_pred, y_test))
print(r2_score(y_test, y_pred))
```

```python
# 라쏘
from sklearn.linear_model import Lasso

model = Lasso(alpha=0.1)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(mean_squared_error(y_pred, y_test))
print(mean_absolute_error(y_pred, y_test))
print(r2_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# decision tree

model = DecisionTreeRegressor(max_depth=50)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(mean_squared_error(y_pred, y_test))
print(mean_absolute_error(y_pred, y_test))
print(r2_score(y_test, y_pred))
```

```python
# random forest
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=114, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(mean_squared_error(y_pred, y_test))
print(mean_absolute_error(y_pred, y_test))
print(r2_score(y_test, y_pred))
```

```python
# random forest
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=10020, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(mean_squared_error(y_pred, y_test))
print(mean_absolute_error(y_pred, y_test))
print(r2_score(y_test, y_pred))
```

```python
# xgboost
import xgboost as xgb
```

```python
model = xgb.XGBRegressor(
    objective ='reg:squarederror',
    n_estimators=1500,
    max_depth=5,
    random_state=42
)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(mean_squared_error(y_pred, y_test))
print(mean_absolute_error(y_pred, y_test))
print(r2_score(y_test, y_pred))
```
